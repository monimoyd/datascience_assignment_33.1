```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

## Load iris dataset

```python
iris = datasets.load_iris() 
```

```python
## Print iris data and target
```

```python
print(iris.data)
```

```python
print(iris.target)
```

## Apply PCA to reduce dimension of IRIS data to 3

iris data has 4 dimensions. Using PCA reducing dimesion to 3


```python
pca = PCA(n_components=3, whiten=True).fit(iris.data)
reduced_iris = pca.transform(iris.data)
print(reduced_iris)
```

## Display reduced iris data in 3 dimensions

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
        
y = np.choose(iris.target, [1, 2, 0]).astype(np.float)

ax.scatter(reduced_iris[:, 0], reduced_iris[:, 1], reduced_iris[:, 2], c=y, cmap=plt.cm.nipy_spectral,
           edgecolor='k')   
    
       
ax.set_xlabel('Setosa')
ax.set_ylabel('Versicolour')
ax.set_zlabel('Virginica')

plt.show()
```
